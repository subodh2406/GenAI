Cosine similarity



```python
# Example 1: Finding similar sentences
sentences = [
    "The cat sat on the mat",
    "A feline rested on the rug",
    "The dog played in the yard",
    "I love programming in Python",
    "Python is my favorite programming language"
]
```

```python
import numpy as np
def cosine_similarity(v1, v2):
    dot_p = np.dot(v1, v2)
    norm_a = np.linalg.norm(v1)
    norm_b = np.linalg.norm(v2)
    return dot_p / (norm_a*norm_b)
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
embeddings
```

```python
sentence_embeddings=embeddings.embed_documents(sentences)
sentence_embeddings
```

```python
# calculating similarity between al the pairs

for i in range(len(sentences)):
    for j in range(i+1, len(sentences)):
        similarity = cosine_similarity(sentence_embeddings[i], sentence_embeddings[j])

        print(f"'{sentences[i]}' vs '{sentences[j]}'")
        print(f"Similarity: {similarity:.3f}\n")
```

Semantic search


```python
### Example- Semantic Search- Retireve the similar sentence
# Test semantic search
documents = [
    "LangChain is a framework for developing applications powered by language models",
    "Python is a high-level programming language",
    "Machine learning is a subset of artificial intelligence",
    "Embeddings convert text into numerical vectors",
    "The weather today is sunny and warm"
]
query="What is Langchain?"
```

```python
def semantic(query, documents, embeddings, top_k=3):

    q_embd = embeddings.embed_query(query)
    d_embd = embeddings.embed_documents(documents)

    similarties = []
    for i, doc_embd in enumerate(d_embd):
        similarity=cosine_similarity(q_embd,doc_embd)
        similarties.append((similarity,documents[i]))

    ## Sort by similarity
    similarties.sort(reverse=True)
    return similarties[:top_k]
```

```python
results=semantic(query,documents,embeddings)
results
```

```python
print(f"\n Semantic Search Results for: '{query}'")
for score, doc in results:
    print(f"Score: {score:.3f} | {doc}")
```
